**SUB PLOT**

```python
X=[10,5,20,15,35]
plt.subplot(2,2,1)
plt.plot(X,color='red',linestyle='-',linewidth=2,marker='o',label='Line')
plt.legend()

plt.subplot(2,2,2)
plt.plot(X,color='green',linestyle='-',linewidth=2,marker='o',label='Line')
plt.legend()

plt.subplot(2,2,3)
plt.plot(X,color='blue',linestyle='-',linewidth=2,marker='o',label='Line')
plt.legend()

plt.subplot(2,2,4)
plt.plot(X,color='black',linestyle='-',linewidth=2,marker='o',label='Line')
plt.legend()

```
